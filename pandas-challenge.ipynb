```python
import pandas as pd
```

```python
#Reading in the Schools csv file
school_summary="PyCitySchools/schools_complete_ csv.csv"
school_summary_df = pd.read_csv(school_summary)
school_summary_df.head()
```

```python
#Reading in the Students csv file
student_summary="PyCitySchools/students_complete_csv.csv"
student_summary_df = pd.read_csv(student_summary)
student_summary_df.head()
```

```python
#DISTRICT SUMMARY:
#Total Schools
school_count=len(school_summary_df["school_name"])
print(school_count)
```

```python
#Total Students
student_count=sum(school_summary_df["size"])
print(student_count)
```

```python
#Total Budget
total_budget=sum(school_summary_df["budget"])
print(total_budget)
```

```python
#Average Math Score
student_summary_df["math_score"].mean(axis=0)
```

```python
#Average Reading Score
student_summary_df["reading_score"].mean(axis=0)
```

```python
#% Passing Math (The percentage of students that passed math.)
    #Made reduced df w/ students who passed math and found length then divided by student_count
    #Matches answer key if we consider passing to be >69%
passing_math_df = student_summary_df.loc[(student_summary_df["math_score"] > 69)]
passing_math_count=len(passing_math_df["math_score"])
print(passing_math_count)
passing_math_percent= (passing_math_count / student_count)*100
print(passing_math_percent)
passing_math_df.head()
```

```python
#% Passing Reading (The percentage of students that passed reading.)
    #Made reduced df w/ students who passed reading and found length then divided by student_count
    #Matches answer key if we consider passing to be >69%
passing_reading_df = student_summary_df.loc[(student_summary_df["reading_score"] > 69)]
passing_reading_count=len(passing_reading_df["reading_score"])
print(passing_reading_count)
passing_reading_percent= (passing_reading_count / student_count)*100
print(passing_reading_percent)
passing_reading_df.head()
```

```python
#% Overall Passing (The percentage of students that passed math and reading.)
passing_overall_df = student_summary_df.loc[(student_summary_df["reading_score"] > 69) & (student_summary_df["math_score"] > 69)]
##Doesn't matter in the next row if we put math or reading score in "" because if one is true both are bc of the &
passing_overall_count=len(passing_overall_df["math_score"])
print(passing_overall_count)
passing_overall_percent= (passing_overall_count / student_count)*100
print(passing_overall_percent)
passing_overall_df.head()
```
